## 1 - Criando uma lista com os nomes de todos os munic√≠pios do Cear√°

````python
# Site com todos os munic√≠pios do Cear√°
lista_munic = "https://www.ipece.ce.gov.br/consulta-aos-mapas-e-legislacao/"
````

````python
# Fazendo a requisi√ß√£o do site por meio do Requests
resposta = requests.get(lista_munic)
````

````python
print(resposta)
````

````python
# Parsear o HTML da p√°gina com a biblioteca BeautifulSoup
municipios = BeautifulSoup(resposta.text, 'html.parser') 
````

````python
# Mostrando o HTML organizado
print(municipios.prettify())
````

````python
# Selecionando a tag da p√°gina HTML que cont√©m o conte√∫do que queremos
municipios = municipios.find_all("strong")
municipios
````

````python
municipios[0].contents
````

````python
len(municipios)
````

````python
# Fazendo um for para pegarmos todos os nomes dos munic√≠pios e armazen√°-los na vari√°vel munic_ceara
munic_ceara = []
for i in range(len(municipios)):
    munic_ceara.append(municipios[i].text)
````

````python
print(munic_ceara)
len(munic_ceara)
````

**Observa√ß√£o:** para a cria√ß√£o da lista com a URL de cada munic√≠pio ser√° necess√°rio removermos os acentos e substituir os espa√ßos em branco por h√≠fens. 

````python
# remover acentos de cada munic√≠pio na lista
municipios_sem_acentos = [unidecode(municipio) for municipio in munic_ceara]

print(municipios_sem_acentos)
len(municipios_sem_acentos)
````

````python
municipios_sem_espacos = [municipio.replace(' ', '') if ' ' in municipio else municipio for municipio in municipios_sem_acentos]

print(municipios_sem_espacos)
````

````python
municipios_com_hifen = [municipio.replace(' ', '-') if ' ' in municipio else municipio for municipio in municipios_sem_acentos]
print(municipios_com_hifen)
````

## 3 - Fazendo o scraping

------------
A estrutura do html onde esta os valores que eu quero fazer o scrapy, √© a seguinte:

```
<div>
    <div class="ind-label">
      <img alt="√Årea Territorial" >
      <p>√Årea Territorial</p>
    </div>
    <p class="ind-value">
      text
    </p>
</div>
```
Para eu poder conseguir pegar o texto que esta na segunda tag 'p', eu vou procurar por uma tag 'img', e pegar a tag antecessora a ela usando o metodo '.find_previous()', mais ainda n√£o √© a tag que eu quero, ent√£o vou usar novamento o metodo 'find_previous()', agora a tag que vai ser retornada vai ser a primeira div (a div pai), pronto agora √© s√≥ pegar o texto da tag 'p' 



````python
headers = { 
    'User-Agent'      : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36', 
    'Accept'          : 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8', 
    'Accept-Language' : 'en-US,en;q=0.5',
    'DNT'             : '1', # Do Not Track Request Header 
    'Connection'      : 'close'
}

results_list = []
counter = 1
````

````python
'indicador unidade indicador unidade'.split(' ')
````

````python
#data = requests.get(url, headers=headers, timeout=5).text
# por meio do requests, 
````

````python
for url in urls:
  try:
    data = requests.get(url, headers=headers, timeout=5).text
    print(f'Requisi√ß√£o {counter}: {url}')
    soup = BeautifulSoup(data,"html.parser")
    

    # Vamos filtrar os valores desejados e salv√°-los detro de um dicionario que vai ser salvo na lista results_list
    codigo = soup.find('div',class_='quick-facts-titulo').find('p',class_='codigo').text.split(' ')[1]
    cidade = soup.find('div',class_='quick-facts-titulo').find('h1').text
    prefeito = soup.find('p', text="Prefeito").find_previous().find_previous('div').find('p',class_='ind-value').text.split('\xa0\xa0\xa0')[0]
    gentilico = soup.find('p', text="Gent√≠lico").find_previous().find_previous('div').find('p',class_='ind-value').text.split('\xa0\xa0\xa0')[0]
    area_territorial = soup.find('img', alt='√Årea Territorial').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    
    
    populacao_estimada = soup.find('img', alt='Popula√ß√£o estimada').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    densidade_demografica = soup.find('img', alt='Densidade demogr√°fica').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    escolarizacao = soup.find('img', alt='Escolariza√ß√£o <span>6 a 14 anos</span>').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    idhm = soup.find('img', alt='IDHM <span>√çndice de desenvolvimento humano municipal</span>').find_previous().find_previous('div').find('p',class_='ind-value').text[:5]
    mortalidade_infantil = soup.find('img', alt='Mortalidade infantil').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    receitas_realizadas = soup.find('img', alt='Receitas realizadas').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    despesas_empenhadas = soup.find('img', alt='Despesas empenhadas').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    pib = soup.find('img', alt='PIB per capita').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    
    cidade_x = {
        'codigo': codigo,
        'cidade': cidade,
        'prefeito': prefeito,
        'gentilico': gentilico,
        'area_territorial': area_territorial,
        'populacao_estimada': populacao_estimada,
        'densidade_demografica': densidade_demografica,
        'escolarizacao': escolarizacao,
        'idhm': idhm,
        'mortalidade_infantil': mortalidade_infantil,
        'receitas_realizadas': receitas_realizadas,
        'despesas_empenhadas': despesas_empenhadas,
        'pib': pib
        }
    results_list.append(cidade_x)

    counter += 1
    #Vai espera um tempo aleat√≥rio entre 4 e 5 segundos antes de fazer a pr√≥xima requisi√ß√£o
    time.sleep(random.uniform(4, 5))
  except requests.exceptions.RequestException as e:
        print("Erro ao fazer a requisi√ß√£o:", e)
        counter += 1
````

````python
#Trasformado a lista 'results_list' em um dataframe
df_cidade = pd.DataFrame(results_list)
df_cidade.head()
````

````python
#Salvado o dataframe
df_cidade.to_csv('DF_Cidades.csv',index=False)
````
